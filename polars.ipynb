# Same filter With Pandas

```python
import pandas as pd

df = pd.read_csv("https://j.mp/iriscsv")
df.query('sepal_length > 5') \
    .groupby('species').sum()
```

# Lazy Example

```python
(pl.read_csv("https://j.mp/iriscsv")
 .lazy()
 .filter(pl.col('sepal_length') > 5)
 .groupby('species', maintain_order=True)
 .agg(pl.all().sum())
 .collect()
)
```

# Filtering

```python
flights = pl.read_parquet('flights.parquet')
```

```python
flights.sample(2)
```

```python
flights.filter(
    pl.col('time_hour').is_between(datetime(2013,9,1), datetime(2013,10,1))
).head()
```

```python
flights.filter(
    (pl.col('dep_time') > 10) & (pl.col('dep_delay') > 10 )
).head()
```

## Creating New Columns

```python
flights.with_columns([
    pl.col('dep_time').mean().alias('average_det_time'),
    (pl.col('dep_delay') > 15).alias('long_dep_delay')
]
).head()
```

```python
flights.groupby('origin', maintain_order=True).agg([
    pl.col('dep_time').mean().alias('average_dep_time'),
    pl.col('dep_delay').max().alias('max_delay')
])
```

## Combining Dataframe

```python
import numpy as np
from datetime import datetime, timedelta
```

```python
df = pl.DataFrame({
    'a': np.arange(0, 8),
    'b': np.random.rand(8),
    'c': [datetime(2022, 12, 1) + timedelta(days=idx) for idx in range(8)],
    'd': [1, 2.0, np.NaN, np.NaN, 0, -5, -42, None]
})

df2 = pl.DataFrame({
    'x': np.arange(0, 8),
    'y': ['A', 'A', 'A', 'B', 'B', 'C', 'X', 'X']
})
```

```python
df.join(df2, left_on='a', right_on='x')
```

```python
pl.concat([df, df2], how='horizontal')
```
